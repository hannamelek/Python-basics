```python
def count_letter(sequence,base):
    count=0
    
    for letter in sequence:
        if letter == base:
            count= count + 1
            
    return count
```

```python
print(count_letter("ATCGATCG","G"))
```

```python
print(count_letter("ATCGATCG", "A"))
```

```python
print(count_letter("ATCGATCG", "T"))
```

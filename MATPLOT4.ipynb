```python
import matplotlib.pyplot as plt

dna = "ATCGATCGAAA"

counts = {
    "A": 0,
    "T": 0,
    "C": 0,
    "G": 0
}

for letter in dna:
    counts[letter] = counts[letter] + 1

bases = counts.keys()
values = counts.values()

plt.bar(bases, values)

plt.xlabel("Nucleotide")
plt.ylabel("Count")
plt.title("DNA Base Frequencies")

plt.show()
```

```python
import matplotlib.pyplot as plt

dna = "GGGGGTTTAAC"

counts = {
    "A": 0,
    "T": 0,
    "C": 0,
    "G": 0
}

for letter in dna:
    counts[letter] = counts[letter] + 1

bases = counts.keys()
values = counts.values()

plt.bar(bases, values)

plt.xlabel("Nucleotide")
plt.ylabel("Count")
plt.title("DNA Base Frequencies")

plt.show()
```

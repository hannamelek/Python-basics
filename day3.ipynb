```python
file = open("sample.fasta")

sequences= []

for line in file:
    line = line.strip()

    if line.startswith(">"):
        continue

    sequences.append(line)

print(sequences)




    
```

```python
output = open("sequences.txt","w")

for seq in sequences:
    output.write(seq + "\n")

output.close()
```

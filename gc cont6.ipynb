```python
dna = "ATGCATGCC"
```

```python
gc_count = 0
```

```python
for letter in dna:
    
    if letter == "G" or letter == "C":
        gc_count = gc_count + 1

gc_percent= (gc_count / len(dna)) * 100

print(gc_percent)
```

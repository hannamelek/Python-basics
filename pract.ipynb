```python
# Count nucleotides in a DNA string
dna = "AAAACCCGGT"

A = 0
C = 0
G = 0
T = 0

for letter in dna:

    if letter == "A":
        A = A + 1

    elif letter == "C":
        C = C + 1

    elif letter == "G":
        G = G + 1

    elif letter == "T":
        T = T + 1

print(A, C, G, T)
```

```python
#REVC reverse complement of DNA
```

```python
dna = "ATTGCCGATTGCC"

complement = ""

for letter in dna:

    if letter == "A":
        complement = complement + "T"

    elif letter == "T":
        complement = complement + "A"
      
    elif letter == "C":
        complement = complement + "G"

    elif letter == "G":
        complement = complement + "C"

reverse = complement[::-1]

print(reverse)
```

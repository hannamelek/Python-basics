```python
s1 = "GAGCCTACTAACGGGAT"
s2 = "CATCGTAATGACGGCCT"
```

```python
distance = 0
```

```python
for i in range(len(s1)):

    if s1[i] != s2[i]:
       distance = distance + 1

print(distance)
    
```

```python
k = 2
m = 2
n = 2

total = k + m + n

recessive = (
    (n/total) * ((n-1)/(total-1)) +
    (n/total) * (m/(total-1)) * 0.5 +
    (m/total) * (n/(total-1)) * 0.5 +
    (m/total) * ((m-1)/(total-1)) * 0.25
)

dominant = 1 - recessive

print(dominant)
```
